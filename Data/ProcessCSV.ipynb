```python
import pandas as pd

# Specify the Excel file path
file_path = 'real_estate_prices.xlsx'
# Specify the output CSV file path
output_csv_path = 'last_rows.csv'

# Read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(file_path, sheet_name=None)

# List to hold the last rows
last_rows = []

# Iterate through each sheet
for sheet_name, df in dfs.items():
    # Get the last row of the DataFrame
    last_row = df.iloc[-1]  # iloc[-1] to access the last row
    # Convert to DataFrame to maintain structure
    last_rows.append(last_row)

# Create a DataFrame from the last rows
last_rows_df = pd.DataFrame(last_rows)

# add a column to indicate the sheet name if needed
last_rows_df['Sheet Name'] = dfs.keys()

# Save the DataFrame containing the last rows to a CSV file
last_rows_df.to_csv(output_csv_path, index=False)

print(f"Last rows extracted and saved to {output_csv_path}")
```

```python
import pandas as pd

# Specify the Excel file path
file_path = 'real_estate_prices.xlsx'
# Specify the output CSV file path
output_csv_path = 'real_estate_prices_all_years.csv'

# Read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(file_path, sheet_name=None)

# List to hold all rows
all_rows = []

# Iterate through each sheet
for sheet_name, df in dfs.items():
    # Add a new column for the sheet name
    df['Sheet Name'] = sheet_name  # Add a column to identify the source sheet
    all_rows.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(all_rows, ignore_index=True)

# Save the combined DataFrame to a CSV file
combined_df.to_csv(output_csv_path, index=False)

print(f"All rows extracted and saved to {output_csv_path}")
```
